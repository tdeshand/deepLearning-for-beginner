```python
import tensorflow.keras 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

<h1 style='color:skyblue;'>Basic linear regression example with keras</h1>

```python
#make data set
x_linear = np.linspace(-10, 10, 100)

y_linear = 2 * x_linear + 5

plt.plot(x_linear, y_linear)
plt.grid(True)
```

```python
#get instance from sequential model
model = Sequential()
```

```python
#add layers to model
#- Dense is a fully connected layer
#-- 1st parameter "units" in Dense layer means how many output 
#-- input_dim means how many dimentions in input
model.add(Dense(1, input_dim=1))

#-- metrics -> what details should be display when the traning happen
model.compile(optimizer=Adam(lr=0.01), loss='mse', metrics=['accuracy'])
```

```python
#train the model
#-- verbose -> how should be displayed metrics
model.fit(x_linear, y_linear, epochs=1000, verbose=2)
```

```python
y_pred = model.predict(x_linear)
```

```python
plt.plot(x_linear, y_linear)
plt.plot(x_linear, y_pred)

plt.legend(['real', 'predict'])
```

```python
loss_log = model.history.history['loss']
plt.plot(loss_log)

plt.ylabel('Loss')
```

<h1 style='color:skyblue;'>Basic logistic regression example with keras</h1>
<h2 style='color:yellowgreen;'>Predicting "logical AND" gate </h2>

```python
x_in = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y_and = np.array([0, 0, 0, 1]).reshape(-1, 1)
```

```python
modelAnd = Sequential()

# In default activation function is None
# input_dim = 2; because here each row has 2 values
modelAnd.add(Dense(units = 1, input_dim = 2, activation='sigmoid'))

#-- reason for using "sigmoid" func is output must be in 0 & 1

modelAnd.compile(optimizer=Adam(lr=0.1), loss=tensorflow.keras.losses.binary_crossentropy, metrics=['accuracy'])
```

```python
modelAnd.fit(x_in, y_and, epochs=500, verbose=2)
```

```python
plt.plot(modelAnd.history.history['loss'])
plt.plot(modelAnd.history.history['acc'])

plt.legend(['loss', 'accuracy'])
plt.xlabel('epochs')
```

```python
modelAnd.predict(np.array([[1, 1]]))
```
